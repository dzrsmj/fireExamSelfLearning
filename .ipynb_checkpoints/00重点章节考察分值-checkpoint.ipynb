# 第二篇 建筑防火
1. 建筑的火灾危险性分类：2-4分 （**8h**）
    1. [生产火灾危险性分类及列举表（表2-2-1 掌握注释内容）](生产火灾危险性分类.ipynb)
    2. 不按物质火灾危险特性确定生产火灾危险性类别的最大允许量表（表2-2-2，甲类12）
    3. 储存物品的火灾危险性分类及举例表（表2-2-3 掌握注释内容）
    4. 典型场所的火灾危险性风险
2. 建筑分类和耐火等级：4-6分 (**10h**)
    1. 民用建筑的分类
    2. 建筑高度的检查内容
    3. 建筑材料燃烧性能等级的附加信息和标识
    4. 影响耐火极限的要素
    5. 厂房和仓库、民用建筑的耐火等级表（表2-3-7和表2-3-8）
3. 总平面图布局和总平面图布置：3-7分 （**16h**）
    1. 掌握建筑选址、城市总体布局、常见企业总平面图布局
    2. 掌握厂房之间及其与乙、丙丁、戊类仓库、民用建筑的防火间距表2-4-1
    3. 掌握民用建筑之间的防火间距表2-4-4（注意放宽和从严的规定）
    4. 熟悉防火间距不足时的消防技术措施
    5. 掌握不同场所的建筑平面布置
    6. 掌握救援设施的布置
    7. 熟悉消防车道的检查内容
4. 防火防烟分区与分隔：3-5分 （**14h**）
    1. 熟悉厂房的层数和每个防火分区的最大允许建筑面积
    2. 熟悉仓库的层数和面积
    3. 掌握不同耐火等级民用建筑防火分区的最大允许建筑面积表2-5-3
    4. 掌握功能区域分隔
    5. 熟悉设备用房分隔
    6. 掌握中庭防火分隔
    7. 熟悉有顶棚的步行街
    8. 掌握防火分隔设施与措施
    9. 熟悉防烟分区的面积划分和分隔措施
    10. 掌握防火阀、排烟防火阀和防火隔间设置
5. 安全疏散：8-12分 （**15h**）
    1. 掌握人员密度的计算
    2. 掌握疏散宽度和疏散距离指标
    3. 熟悉安全出口与疏散出口的设置要求（检查内容）
    4. 熟悉疏散走道设置的基本要求
    5. 掌握避难走道的设置要求
    6. 熟悉疏散楼梯的检查内容
    7. 掌握防烟楼梯间的使用范围和设置要求
    8. 掌握避难层（间）设置要点
6. 建筑电气防火：2-3分 （**2h**）
    1. 熟悉电气火灾隐患特征
    2. 掌握用电设备防火，尤其是照明器具防火，主要为常识性内容积累
7. 建筑防爆：3-4分 （**3h**）
    1. 熟悉建筑防爆原则和措施
    2. 掌握爆炸危险性厂房、库房的布置要求
    3. 掌握爆炸危险环境中电气的防爆措施

# 第三篇 建筑消防设施
1. 消防给水及消火栓系统：6-10分 （**13h**）
    1. 重点掌握消防给水系统的构成、系统类型
    2. 消防水泵设置要求
    3. 消防水泵接合器设置要求
    4. 消防水池、水箱的设置要求
    5. 室内外消火栓的设置要求和范围
2. 自动喷水灭火系统：10分 （**15**）
    1. 重点掌握系统分类和组成、各系统的工作原理和适用范围
    2. 自喷设置场所的火灾危险性举例要牢记
    3. 系统主要组件的设置要求
3. 水喷雾和细水雾灭火系统：4分 （**8h**）
    1. 重点掌握水喷雾和细水雾系统灭火机理、分类与适用范围
    2. 注意两者的区别
4. 自动跟踪定位射流灭火系统：3分 （**5h**）
    1. 熟悉系统分类、组成、工作原理及适用范围
    2. 掌握系统选型、主要组件设置要求
5. 气体灭火系统：5-7分 （**10h**）
    1. 掌握气体灭火系统的组成、分类、控制方式、适用范围及设置要求等
    2. 气体灭火系统的灭火机理、工作原理、防护区的设置
6. 泡沫灭火系统：2-3分 （**5h**）
    1. 重点掌握泡沫灭火系统的使用范围、系统和组件的设置要求
7. 干粉灭火系统：1分 （**1h**）
    1. 掌握干粉灭火系统的组成和分类、适用范围、系统和组件设置要求
8. 火灾自动报警系统：10-14分 （**15h**）
    1. 重点掌握火灾报警系统的分类
    2. 报警区域和探测区域的划分
    3. 火灾探测器的选择
    4. 消防联动控制的设计要求
    5. 消防控制室等重点内容
9. 防排烟系统：4-5分 （**3.5h**）
    1. 重点掌握自然通风和自然排烟方式的选择
    2. 机械加压系统和机械排烟系统的选择和主要设计参数
    3. 防排烟系统的联动控制方式
10. 消防应急照明和疏散指示系统：2-3分 （**1.5h**）  高性价比章节，容易得分
    1. 重点掌握系统的分类和组成
    2. 系统的性能和设置要求
11. 建筑灭火器配置：3-4分 （**4h**）
    1. 重点掌握灭火器的分类、适用范围、场所的危险登记局配置要求
12. 消防供配电：2分 （**1.5h**）
    1. 重点掌握消防用电的负荷等级
    2. 消防备用电源的设计要求和电线电缆的选择
